```python
import sklearn
import numpy as np
import scipy as sp
import matplotlib as mpl
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing
import statsmodels.api as sm
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('Salary_dataset.csv')
print(df.head())
X=np.array(df['YearsExperience'])
y=np.array(df['Salary'])
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=0)
```

```python
X_train=X_train.reshape(-1,1)
X_test = X_test.reshape(-1,1)
lr=LinearRegression().fit(X_train, y_train)
print(f"Linear Regression-Training set score: {lr.score(X_train, y_train):.2f}")
print(f"Linear Regression-Test set score: {lr.score(X_test, y_test):.2f}")
```

```python
a=lr.coef_[0]
b=lr.intercept_
print(a, "* x +", b, "= y")
```

```python
lr.predict([[20]])
```

```python
a*20+b
```

```python
pred = lr.predict([[20]])
fig_scat, ax_scat = plt.subplots(1,1,figsize=(10,6))
x = X.T
best_fit = b + a * (np.append(x,[20]))
ax_scat.scatter(x,y,s=300, label="Training Data")
ax_scat.plot(np.append(x,[20]),best_fit, ls='--', label="Best Fit Line")
ax_scat.plot([20],pred,"ys",label="LinearRegression")
ax_scat.set_xlabel(r'$x$')
ax_scat.set_ylabel(r'$y$')
```

```python
df=pd.read_csv('score.csv')
print(df.head())
X2=np.array(df['Hours'])
y2=np.array(df['Scores'])
X2_train, X2_test, y2_train, y2_test = train_test_split(X2,y2,test_size=0.25, random_state=0)
X2_train=X2_train.reshape(-1,1)
X2_test = X2_test.reshape(-1,1)
lr=LinearRegression().fit(X2_train, y2_train)
print(f"Linear Regression-Training set score: {lr.score(X2_train, y2_train):.2f}")
print(f"Linear Regression-Test set score: {lr.score(X2_test, y2_test):.2f}")
```

```python
a=lr.coef_[0]
b=lr.intercept_
print(a, "* x +", b, "= y")
```

```python
# Assuming x_train, y_train, x_test, y_test are your training and test data
# Assuming b, a are the intercept and slope of the best fit line
# Assuming pred is the prediction made by the Linear Regression model

# Create a figure and axis for scatter plot
fig_scat, ax_scat = plt.subplots(1, 1, figsize=(10, 6))

# Scatter plot for training data
ax_scat.scatter(X2_train, y2_train, color='blue', label="Training Data")

# Scatter plot for test data
ax_scat.scatter(X2_test, y2_test, color='red', label="Test Data")

# Plot the best fit line
x = np.linspace(min(X2_train), max(X2_train), 100)
best_fit = b + a * x
ax_scat.plot(x, best_fit, color='green', linestyle='--', label="Best Fit Line")

# Plot the prediction made by the Linear Regression model
ax_scat.plot([20], pred, "ys", label="Linear Regression Prediction")

# Set labels and title
ax_scat.set_xlabel('x')
ax_scat.set_ylabel('y')
ax_scat.set_title('Linear Regression: Training and Test Data')

# Add legend
ax_scat.legend()

# Show plot
plt.show()
```
